```python
import os
import cv2
import pytesseract
import re
import pandas as pd
from pdf2image import convert_from_path
from google.colab import files, drive
```

```python
use_drive = True  # Set to True if using Google Drive
if use_drive:
    file_path = "/content/VietjetAir.pdf"
else:
    uploaded = files.upload()
    file_path = list(uploaded.keys())[0]

images = convert_from_path(file_path, dpi=300)
image_paths = []
for i, img in enumerate(images):
    img_path = f"page_{i}.png"
    img.save(img_path, "PNG")
    image_paths.append(img_path)

def preprocess_image(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    processed = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
    cv2.imwrite(image_path.replace(".png", "_processed.png"), processed)
    return image_path.replace(".png", "_processed.png")

processed_images = [preprocess_image(img) for img in image_paths]

extracted_text = []
for img in processed_images:
    text = pytesseract.image_to_string(img, lang="eng")
    extracted_text.append(text)

financial_data = []
for text in extracted_text:
    amounts = re.findall(r"\$?\d{1,3}(?:,\d{3})*(?:\.\d{2})?", text)
    financial_data.extend(amounts)

df_text = pd.DataFrame({"Extracted_Text": extracted_text})
df_text.to_csv("financial_report_text.csv", index=False)

df_financial = pd.DataFrame({"Financial_Figures": financial_data})
df_financial.to_csv("financial_report_numbers.csv", index=False)

print("Extracted Financial Figures:", financial_data)
```
